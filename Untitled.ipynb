```python
import numpy as np
import json 
import pandas as pd
```

```python
def parse_data(data_file):
    """This method reads json data file and then create a Pandas DataFrame"""
    geo_fence = []
    name_of_date = []
    month = []
    day_number = []
    time_hours = []
    time_minutes = []
    waiting_time = []    
    with open(data_file) as json_file:
        json_data = json.load(json_file)
        for json_object in json_data:
            try:
                geo_fence.append(json_object['values']['geoFence'])
                name_of_date.append(json_object['values']['day'])
                day_number.append(json_object['values']['date'])
                time_hours.append(json_object['values']['time_hours'])
                time_minutes.append(json_object['values']['time_minutes'])
                month.append(json_object['values']['month'])
                waiting_time.append(json_object['values']['waitingTime'])
            except Exception as ex:
                print 'error in processing json string %s, error: %s\n' %(json_object, ex)
                
    #buding DataFrame
    data_frame = pd.DataFrame({
            'geo_fence': geo_fence,
            'day': name_of_date,
            'date': day_number,
            'time_hours': time_hours,
            'time_minutes': time_minutes,
            'month': month,
            'waiting_time':waiting_time
        })
    return data_frame
```

```python
waiting_time = parse_data('./data/big_time_series.json')
```
